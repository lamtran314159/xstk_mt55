```r
grade <- read.csv("All_GPUs.csv", header = TRUE)
head(grade)
```

```r
# Chon cac cot quan trong
new_DF <- grade[, c("Name", "Best_Resolution", "Core_Speed", "Memory","Memory_Speed","Memory_Bandwidth", "Manufacturer", "Release_Date")]
head(new_DF)
```

```r
#Loai bo don vi "MB" tu cot "Memory"
new_DF$Memory <- as.numeric(sub(" MB", "", new_DF$Memory))
#Loai bo don vi "GB/sec" tu cot "Memory_Bandwidth"
new_DF$Memory_Bandwidth <- as.numeric(gsub("\\s*GB/sec", "", new_DF$Memory_Bandwidth))
#Loai bo don vi "MHz" tu cot "Memory_Speed"
new_DF$Memory_Speed <- as.numeric(sub(" MHz", "", new_DF$Memory_Speed))
head(new_DF)
```

```r
#Kiem tra cac du lieu bi khuyet va de xuat cac phuong an
#Kiem tra bao nhieu du lieu bi khuyet
apply(is.na(new_DF),2,sum)
#Kiem tra vi tri chinh xac cua du lieu bi khuyet
apply(is.na(new_DF),2,which)
#tinh ti le du lieu khuyet
apply(is.na(new_DF),2,mean)
```

```r
#Xoa cac du lieu bi khuyet cua cot "Memory\_Speed","Memory\_Bandwidth"
new_DF <- subset(new_DF, 
            !is.na(Memory_Speed) & !is.na(Memory_Bandwidth))
# Tinh median cua cot "Memory" (bo qua gia tri NA)
median_memory <- median(new_DF$Memory, na.rm = TRUE)
# Thay the gia tri NA trong cot "Memory" bang median
new_DF$Memory <- ifelse(is.na(new_DF$Memory), median_memory,new_DF$Memory)
#Kiem tra lai du lieu
apply(is.na(new_DF),2,sum)
```

```r
library(dplyr)
library(knitr)
nF_summ <- new_DF %>%
  group_by(Manufacturer) %>%
  summarize(
    sample_size = n(),
    mean = mean(Memory_Speed, na.rm = TRUE),
    sd = sd(Memory_Speed, na.rm = TRUE),
    minimum = min(Memory_Speed, na.rm = TRUE),
    first_quantile = quantile(Memory_Speed, 0.25, na.rm = TRUE),
    median = median(Memory_Speed, na.rm = TRUE),
    third_quantile = quantile(Memory_Speed, 0.75, na.rm = TRUE),
    maximum = max(Memory_Speed, na.rm = TRUE)
  )
```

```r
library(ggplot2)
ggplot(new_DF, aes(x=Manufacturer, y=Memory_Speed)) +
geom_boxplot() +
stat_summary(fun.y = "mean", geom = "point", color = "red") +
theme_minimal()
```

```r
#đọc dữ liệu 2 file vào
gpus <- read.csv("All_GPUs.csv", header = TRUE)

str(gpus) #in cấu trúc của data (kiểu dữ liệu và thành phần)
#tạo data mới chỉ gồm các cột cần dùng
df_gpus <- gpus[,c("Manufacturer","Open_GL","Shader","Max_Power","Memory","Memory_Bandwidth","Memory_Bus","Memory_Speed")]
head(df_gpus) #in dữ liệu 6 dòng đầu

apply(is.na(df_gpus), 2, sum) #kiểm tra dữ liệu NA của từng cột trong data

print("Xử lý số liệu:")
#GPUs
df_gpus$Max_Power <- gsub(" Watts", '', df_gpus$Max_Power)
df_gpus$Max_Power <- as.numeric(df_gpus$Max_Power)
df_gpus$Memory <- gsub(" MB", '', df_gpus$Memory) #dùng để thay thế ký tự này thành hý tự khác
df_gpus$Memory <- as.numeric(df_gpus$Memory) #ép kiểu dữ liệu thành kiểu số (num)
df_gpus$Memory_Bandwidth <- gsub("GB/sec", '', df_gpus$Memory_Bandwidth)
df_gpus$Memory_Bandwidth <- as.numeric(df_gpus$Memory_Bandwidth)
df_gpus$Memory_Bus <- gsub(" Bit", '', df_gpus$Memory_Bus)
df_gpus$Memory_Bus <- as.numeric(df_gpus$Memory_Bus)
df_gpus$Memory_Speed <- gsub(" MHz", '', df_gpus$Memory_Speed)
df_gpus$Memory_Speed <- as.numeric(df_gpus$Memory_Speed)
#thay các dữ liệu NA bằng giá trị trung bình
df_gpus$Max_Power[is.na(df_gpus$Max_Power)] = mean(df_gpus$Max_Power, na.rm = TRUE)
df_gpus$Open_GL[is.na(df_gpus$Open_GL)] = mean(df_gpus$Open_GL, na.rm = TRUE)
df_gpus$Shader[is.na(df_gpus$Shader)] = mean(df_gpus$Shader, na.rm = TRUE)
df_gpus$Memory[is.na(df_gpus$Memory)] = mean(df_gpus$Memory, na.rm = TRUE)
df_gpus$Memory_Bandwidth[is.na(df_gpus$Memory_Bandwidth)] = mean(df_gpus$Memory_Bandwidth, na.rm = TRUE)
df_gpus$Memory_Bus[is.na(df_gpus$Memory_Bus)] = mean(df_gpus$Memory_Bus, na.rm = TRUE)
df_gpus$Memory_Speed[is.na(df_gpus$Memory_Speed)] = mean(df_gpus$Memory_Speed, na.rm = TRUE)

#---------------------------

print("Tìm khoảng tin cậy:")
t.test(df_gpus$Open_GL, conf.level = 0.95)

#---------------------------

print("Kiểm định một mẫu:")
#H0: mu = 3072
#H1: mu < 3072
t.test(df_gpus$Memory, alternative = "less", mu = 3072, paired = FALSE, var.equal = TRUE, conf.level = 0.95)

#---------------------------

print("Kiểm định hai mẫu:")
t.test(df_gpus$Open_GL, df_gpus$Shader, alternative = "two.sided", mu = 0, paired = FALSE, var.equal = TRUE, conf.level = 0.95)

#---------------------------

print("Anova một yếu tố:")
manu <- factor(df_gpus$Manufacturer)
ogl <- df_gpus$Open_GL

table(df_gpus$Manufacturer) #số lượng mẫu từng biến độc lập

#Kiểm định giả thiết:
# - Các mẫu độc lập
# - Biến phụ thuộc là biến liên tục
# - Các nhóm được lấy ra từ tổng thể có phân phối chuẩn hoặc gần chuẩn

#Kiểm định Shapiro-Wilk để biết có phân phối chuẩn hoặc gần chuẩn:
library(nortest)
shapiro.test(ogl[manu=="Nvidia"])
shapiro.test(ogl[manu=="AMD"])
shapiro.test(ogl[manu=="Intel"])
shapiro.test(ogl[manu=="ATI"])

require(car)
leveneTest(ogl~manu) #kiểm tra các nhóm có phương sai đồng nhất

#phân tích anova
av <- aov(ogl~manu, data = df_gpus)
summary(av) #hiển thị kết quả

TukeyHSD(av) #phân tích sự khác nhau giữa các biến độc lập

#---------------------------

print("Hồi quy tuyến tính đơn:")
#vẽ đồ thị hồi quy
require(ggpubr)
ggscatter(data = df_gpus, x = "Memory_Bandwidth", y = "Memory") + geom_smooth(method = "lm", se = F)

#chia tập dữ liệu thành 2 phần:
# 80% dùng để xây dựng mô hình hồi quy
# 20% dùng để đánh giá mức độ chính xác của mô hình hồi quy
s <- sample(seq_len(nrow(df_gpus)), size = floor(0.8*nrow(df_gpus))) #chọn ngẫu nhiên 80% dòng của df_gpus
train <- df_gpus[s, ]
test <- df_gpus[-s, ]

#hồi quy tuyến tính đơn
lm_train <- lm(train$Memory~train$Memory_Bandwidth, data = train)
summary(lm_train)

#kiểm tra với mẫu còn lại  
lm_test <- lm(test$Memory~test$Memory_Bandwidth, data = test)
summary(lm_test)

lm <- lm(df_gpus$Memory~df_gpus$Memory_Bandwidth, data = df_gpus)
plot(lm)

#---------------------------

print("Hồi quy tuyến tính đa biến:")
mlm <- lm(df_gpus$Memory~df_gpus$Memory_Bandwidth+df_gpus$Memory_Bus+df_gpus$Max_Power, data = df_gpus)
summary(mlm)
```
